```python
import numpy as np
import copy
```

```python
def PhasesGenerator(Datemid,Endmid):
    #--------------------------------------------- Préparation de la routine
    Date=Datemid
    x=[0., 0.0625, 0.1875, 0.25, 0.3125, 0.375, 0.4375, 0.5, 0.5625, 0.625, 0.6875, 0.75, 0.8125, 0.875, 0.9375]
    phases=[]               #phases est notre output final, nous allons la remplir via la boucle suivante
    #--------------------------------------------- Générateur de phases
    for i in range(0,15):
        phases.append([])                         #Nous créons une liste vide dans phases afin d'y répertorier toutes les dates correspondant à une phase précise de notre variabilité
        while Date<=Endmid:
            phases[i].append(Date+x[i]*0.407521)
            Date=Date+0.407521
        Date=Datemid                              #Nous réinitialisons Date avec la valeur du mid initiale avant de changer de phase
    return(phases)
```

```python
PG=PhasesGenerator(2458933.278,2458975.322)#Génère toutes les dates de chaque phase entre le 24/03 et le 05/05
```

```python
def PhasesFiltred(PG):
    #--------------------------------------------- Préparation de la routine
    PGc=copy.deepcopy(PG)        #Nous copions la liste de données. Cela évitera à notre boucle de sauter une donnée lorsque l'on retira une valeur
    L = len(PG)                  #Nous notons le nombre d'éléments dans liste pour la lisibilité de la routine
    #--------------------------------------------- Filtres
    for i in range(0,L):         #Nous commençons par sélectionner la phase désirée
        for j in range(0, len(PG[i])):     #Pour chaque j élément de la phase, nous appliquons nos filtres
            if PG[i][j] >= 2458945. and PG[i][j]<= 2458948.:                #Ce filtre ignore les dates du 05/04 -> 08/04 où la lune est trop proche de notre étoile
                PGc[i].remove(PG[i][j])
            elif ((PG[i][j])%1) <= 0.6455 or ((PG[i][j])%1) >= 0.9375:      #Ce filtre nous permet d'éliminer les moments où notre phase est présente en journée (entre 21h30 et 4h30 heure locale)
                PGc[i].remove(PG[i][j])
    return(PGc)
```

```python
PF=PhasesFiltred(PG)
```

```python
def ConverterJD(jul): #Convertisseur pour une date précise
    #--------------------------------------------- Etape préliminaire pour convertir un jour julien en jour grégorien
    Z=int(jul+0.5)
    F=(jul+0.5)%1
    a=int((Z-1867216.25)/36524.25)
    A=Z+1+a-int(a/4)
    B=A+1524
    C=int((B-122.1)/365.25)
    D=int(365.25*C)
    E=int((B-D)/30.6001)
    #--------------------------------------------- Conversion jour
    d=int(B-D-int(30.6001*E)+F)
    #--------------------------------------------- Conversion mois
    if E<14:
        m=E-1
    else:
        m=E-13
    #--------------------------------------------- Conversion année
    if m>2:
        y=C-4716
    else:
        y=C-4715
    #--------------------------------------------- Conversion heure-minute-seconde
    hour=abs(24*F)
    mini=(hour-int(hour))*60
    sec=(mini-int(mini))*60
    return(m,d,y,int(hour),int(mini),round(sec,0))#Nous arrondissons à la seconde près. Les heures et les minutes ne sont pas arrondies à l'entier le proche mais est conservée telle quelle car, respectivement, les minutes sont converties à partir des décimales des heures et les secondes à partir de celles des minutes.

def TimeConverter(PF): #Permet d'appliquer la routine précédente à notre liste de phases
    PFC=copy.deepcopy(PF)
    x=[0., 0.0625, 0.1875, 0.25, 0.3125, 0.375, 0.4375, 0.5, 0.5625, 0.625, 0.6875, 0.75, 0.8125, 0.875, 0.9375]
    for i in range(0,len(PFC)):
        for j in range(0,len(PFC[i])):
            PFC[i][j]=ConverterJD(PF[i][j])
        print("\n Phase:",x[i],"\n",PFC[i])
```
